## 베이즈 추론 프로그램하기

```python
# 0~1 이하의 범위를 1001개로 분할하는 배열 생성
thetas = np.linspace(0, 1, 1001)
thetas
```

```python
# 가능도 함수 생성
likelihood = lambda r: thetas if r else (1 - thetas)
likelihood
```

```python
# 사후 분포를 계산하는 함수 posterior 생성
def posterior(r, prior):
    lp = likelihood(r) * prior
    return lp / lp.sum()
```

```python
# p(theta) 균일 분포로 가정한 사전 분포를 생성한 배열에 적용하기
p = np.array([1 / len(thetas) for _ in thetas])
```

```python
# posterior 함수를 이용해서 베이즈 추론을 시행 (r =1 이 주어졌을 때의 사후 분포를 계산)
p = posterior(1, p)
print(p)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(thetas, p)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

```python
clicks = 2
no_clicks = 38

p = np.array([1 / len(thetas) for theta in thetas])

for _ in range(clicks):
    p = posterior(1, p)

for _ in range(no_clicks):
    p = posterior(0, p)

plt.plot(thetas, p)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

## 통계 모델링을 활용한 반복 모으기 시행

```python
# 이항 분포를 가정한 가능도 함수 정의
likelihood = lambda a, N: thetas ** a * (1 - thetas) ** (N - a)
```

```python
# 사후 분포를 계산하는 posterior 함수 정의
def posterior(a, N, prior):
    lp = likelihood(a, N) * prior
    return lp / lp.sum()
```

```python
# 사전 분포 prior를 정의하고 데이터를 이용해 사후 분포를 시각화
prior = 1 / len(thetas)

plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40, prior), label="Alice - A")
plt.plot(thetas, posterior(4, 50, prior), label="Alice - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280, prior), label="Bob - A")
plt.plot(thetas, posterior(128, 1600, prior), label="Bob - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()
plt.show()
```

## 베타 분포를 활용한 사후 분포 계산

```python
def betaf(alpha, beta):
    numerator = thetas ** (alpha - 1) * (1 - thetas) ** (beta - 1)
    return numerator / numerator.sum()

def posterior(a, N):
    return betaf(a + 1, N - a + 1)
```

```python
plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40), label="Alice - A")
plt.plot(thetas, posterior(4, 50), label="Alice - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280), label="Bob - A")
plt.plot(thetas, posterior(128, 1600), label="Bob - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()
plt.show()
```
