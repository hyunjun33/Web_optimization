## 베이즈 추론 프로그램하기

```python
# 0~1 이하의 범위를 1001개로 분할하는 배열 생성
thetas = np.linspace(0, 1, 1001)
thetas
```

```python
# 가능도 함수 생성
likelihood = lambda r: thetas if r else (1 - thetas)
likelihood
```

```python
# 사후 분포를 계산하는 함수 posterior 생성
def posterior(r, prior):
    lp = likelihood(r) * prior
    return lp / lp.sum()
```

```python
# p(theta) 균일 분포로 가정한 사전 분포를 생성한 배열에 적용하기
p = np.array([1 / len(thetas) for _ in thetas])
```

```python
# posterior 함수를 이용해서 베이즈 추론을 시행 (r =1 이 주어졌을 때의 사후 분포를 계산)
p = posterior(1, p)
print(p)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(thetas, p)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

```python
clicks = 2
no_clicks = 38

p = np.array([1 / len(thetas) for theta in thetas])

for _ in range(clicks):
    p = posterior(1, p)

for _ in range(no_clicks):
    p = posterior(0, p)

plt.plot(thetas, p)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

## 통계 모델링을 활용한 반복 모으기 시행

```python
# 이항 분포를 가정한 가능도 함수 정의
likelihood = lambda a, N: thetas ** a * (1 - thetas) ** (N - a)
```

```python
# 사후 분포를 계산하는 posterior 함수 정의
def posterior(a, N, prior):
    lp = likelihood(a, N) * prior
    return lp / lp.sum()
```

```python
# 사전 분포 prior를 정의하고 데이터를 이용해 사후 분포를 시각화
prior = 1 / len(thetas)

plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40, prior), label="Alice - A")
plt.plot(thetas, posterior(4, 50, prior), label="Alice - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280, prior), label="Bob - A")
plt.plot(thetas, posterior(128, 1600, prior), label="Bob - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()
plt.show()
```

## 베타 분포를 활용한 사후 분포 계산

```python
def betaf(alpha, beta):
    numerator = thetas ** (alpha - 1) * (1 - thetas) ** (beta - 1)
    return numerator / numerator.sum()

def posterior(a, N):
    return betaf(a + 1, N - a + 1)
```

```python
plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40), label="Alice - A")
plt.plot(thetas, posterior(4, 50), label="Alice - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280), label="Bob - A")
plt.plot(thetas, posterior(128, 1600), label="Bob - B")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()
plt.show()
```

## 사후 분포를 이용해 의사 결정 내리기
### HDI를 이용한 가설 검정

```python
# hmv : highest mass values

def hmv(xs, ps, alpha=0.95):
    xps = sorted(zip(xs, ps), key=lambda xps: xps[1], reverse=True)
    xps = np.array(xps)
    xs = xps[:, 0]
    ps = xps[:, 1]
    return np.sort(xs[np.cumsum(ps) <= alpha])
```

```python
# 엘리스의 디자인 안 A에 대한 사후분포

thetas = np.linspace(0, 1, 1001)

def posterior(a, N):
    alpha = a + 1
    beta = N - a + 1
    numerator = thetas ** (alpha - 1) * (1 - thetas) ** (beta - 1)
    return numerator / numerator.sum()

ps = posterior(2, 40)
```

```python
# 이산화된 사후 분포와 정의한 hmv 메서드를 이용해서 얻은 배열을 hm_thetas로 정의한다
# 구간은 값이 몰려있는 한 구간의 최솟값과 최대값을 구간의 양끝으로 하여 HDI를 구한다 (95% 기준)

hm_thetas = hmv(thetas, ps, alpha=0.95)
plt.figure(figsize=(10 ,6))
plt.plot(thetas, ps)
plt.annotate("", xy=(hm_thetas.min(), 0),
             xytext=(hm_thetas.max(), 0),
             arrowprops=dict(color="black", shrinkA=0, shrinkB=0,
                             arrowstyle="<->", linewidth=2)
            )

plt.annotate("%.3f" % hm_thetas.min(), xy=(hm_thetas.min(), 0),
             ha="right", va="bottom")

plt.annotate("%.3f" % hm_thetas.max(), xy=(hm_thetas.max(), 0),
             ha="left", va="bottom")

plt.annotate("95% HDI", xy=(hm_thetas.mean(), 0),
             ha="center", va="bottom")

hm_region = (hm_thetas.min() < thetas) & (thetas < hm_thetas.max())
plt.fill_between(thetas[hm_region], ps[hm_region], 0, alpha=0.3)

plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.xlim(0, 0.3)
plt.tight_layout()
plt.show()
```

```python
def plot_hdi(ps, label):
    hm_thetas = hmv(thetas, ps, 0.95)
    plt.plot(thetas, ps)
    plt.annotate("", xy=(hm_thetas.min(), 0),
                 xytext=(hm_thetas.max(), 0),
                 arrowprops=dict(color="black", shrinkA=0, shrinkB=0,
                                 arrowstyle="<->", linewidth=2)
                )
    plt.annotate("%.3f" % hm_thetas.min(), xy=(hm_thetas.min(), 0),
                 ha="right", va="bottom")
    plt.annotate("%.3f" % hm_thetas.max(), xy=(hm_thetas.max(), 0),
                 ha="left", va="bottom")
    hm_region= (hm_thetas.min() < thetas) & (thetas < hm_thetas.max())
    plt.fill_between(thetas[hm_region], ps[hm_region], 0, alpha=0.3)
    plt.xlim(0, 0.3)
    plt.ylabel(label)
    plt.yticks([])

plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
alice_a = posterior(2, 40)
plot_hdi(alice_a, "Alice A")

plt.subplot(4, 1, 2)
alice_b = posterior(4, 50)
plot_hdi(alice_b, "Alice B")

plt.subplot(4, 1, 3)
bob_a = posterior(64, 1280)
plot_hdi(bob_a, "Bob A")

plt.subplot(4, 1, 4)
bob_b = posterior(128, 1600)
plot_hdi(bob_b, "Bob B")
plt.xlabel(r"$\theta$")
plt.tight_layout()
plt.show()
```

```python
# 앨리스의 보고서 디자인 A안의 클릭률의 사후 분포로부터 얻은 샘플의 히스토그램

data = np.random.beta(3, 39, size=100000)
plt.hist(data, range=(0, 0.3), bins=60)
plt.xlabel(r"$\theta$")
plt.ylabel("Frequency")
plt.show()
```

```python
# 앨리스의 보고서 디자인안의 클릭률 차이를 나타내는 확률 변수 시그마의 히스토그램

theta_a = np.random.beta(3, 39, size=100000)
theta_b = np.random.beta(5, 47, size=100000)
delta = theta_b - theta_a

plt.hist(delta, range=(-0.3, 0.3), bins=60)
plt.xlabel(r"$\delta$")
plt.ylabel("Frequcney")
plt.show()
```

```python
# 클릭률의 차이 시그마를 이용하면 다음의 정량적 평가로 변환이 가능하다
# 정량적 평가 : 시그마가 95%의 확률로 양수가 된다 (theta_b의 값이 더 크다)

print((delta >0).mean()) # 95% 신뢰구간을 기준으로 평가한다는 가정
```
